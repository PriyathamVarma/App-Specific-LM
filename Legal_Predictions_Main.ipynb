```python
import pandas as pd

# Assuming 'df' is your DataFrame
encoded_df = data_frame.copy()  # Create a copy of the original DataFrame

# Flatten the DataFrame into a Series
flattened_series = data_frame.stack()

# Get unique values
unique_values = flattened_series.unique()

# Create a mapping dictionary to assign unique numbers to each value
# Handle NaN by assigning a special value (e.g., -1)
mapping_dict = {value: index if pd.notna(value) else -1 for index, value in enumerate(unique_values)}

# Map each value in the DataFrame to its corresponding unique number using the mapping dictionary
encoded_df = encoded_df.applymap(lambda x: mapping_dict.get(x, -1)) # Handle potential missing values in mapping_dict

# Now, 'encoded_df' contains the entire DataFrame with each value encoded as a unique number, and NaN values replaced with -1
encoded_df
```

# Classifiers

```python
#knn classifier
from sklearn.neighbors import KNeighborsClassifier

# Flatten y_test and y_train
y_test_flat = np.ravel(y_test)
y_train_flat = np.ravel(y_train)

clf_nn = KNeighborsClassifier(n_neighbors=6)
# n_neighbours < test size, so if n_neighbours is 6 then test_size should be < 0.6
# Expected n_neighbors <= n_samples,  but n_samples = 5, n_neighbors = 6
clf_knn=clf_nn
clf_knn.fit(x_train,y_train_flat)
y_hat = clf_knn.predict(x_test)
y_hat


from sklearn.metrics import accuracy_score

#we can use the built in accuracy metrics in sklearn.
accuracy = accuracy_score(y_test, y_hat)
print('Sklearn calculated accuracy:',accuracy)



#We could implement our own accuracy.
#find the indices where the predicted and the actual match
counts, = np.where(y_hat==y_test_flat)

print('Number of data points in test data:',len(y_test))
print('Number of data points where prediction is correct:',len(counts))

manual_accuracy = len(counts) / len(y_test)

print('Manual accuracy:',manual_accuracy)
print(counts)
print('Prediction', y_hat)
print('Actual    ',y_test)
```

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier

# Flatten y_test and y_train
y_test_flat = np.ravel(y_test)
y_train_flat = np.ravel(y_train)

random_variable = RandomForestClassifier(max_depth=10)

random_variable.fit(x_train,y_train)

new_random_variable = random_variable.predict(x_test)

new_random_variable


from sklearn.metrics import accuracy_score

#we can use the built in accuracy metrics in sklearn.
accuracy2 = accuracy_score(y_test, new_random_variable)
print('Sklearn calculated accuracy:',accuracy2)

#We coould implement our own accuracy.
#find the indices where the predicted and the actual match
counts, = np.where(new_random_variable==y_test_flat)

print('Number of data points in test data:',len(y_test))
print('Number of data points where prediction is correct:',len(counts))

manual_accuracy2 = len(counts) / len(y_test)

print('Manual accuracy:',manual_accuracy2)
print(counts)
print('Prediction', new_random_variable)
print('Actual    ',y_test)
```
